```python
import sys
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
fifa_filepath = "data/fifa.csv"

# Read the file into a variable fifa_data
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
```

```python
memory_filepath = "data/memory.csv"
memory_data = pd.read_csv(memory_filepath)


plt.figure(figsize = [20,10],dpi=100)
sns.set(style='white')
ax = sns.barplot(
    data = memory_data,
    x= memory_data['Benchmarks'], 
    y= memory_data['Memory Usage (m)'], 
    hue=memory_data['JVM']
    )
```

```python
time_filepath = "data/execution_time.csv"
time_data = pd.read_csv(time_filepath)


plt.figure(figsize = [20,10],dpi=100)
sns.set(style='white')
ax = sns.barplot(
    data = time_data,
    x= time_data['Benchmarks'], 
    y= time_data[' Execution Time (s)'], 
    hue=time_data['JVM']
    )
```

```python
throughput_filepath = "data/throughput.csv"
throughput_data = pd.read_csv(throughput_filepath)


plt.figure(figsize = [20,10],dpi=100)
sns.set(style='white')
ax = sns.barplot(
    data = throughput_data,
    x= throughput_data['Benchmarks'], 
    y= throughput_data['Throughput (%)'], 
    hue=throughput_data['JVM']
    )
```
